```python
import pandas as pd
import numpy as np 
```

```python
# load in the current dataset 
dataset = pd.read_csv("predictingCSectionsInWomanData.csv", header=0)

# drop the unnecessary rows and columns 
dataset.drop(dataset.columns[[18, 19, 20, 21]], axis=1, inplace=True)
dataset.drop(dataset.index[[302, 303, 304,305, 306]], inplace=True)

dataset
```

```python
# converting the Delivery to numerical values 
# 0 --> Delivery 
# 1 --> Cesarean 

dataset = dataset.replace({"Delivery": 0, "Cesarean":1})
features = list(dataset.columns)

print(features)

dataset
```

```python
# creating the first decision tree 

from sklearn import tree 

# convert dataset to numpy for easier usse
data = dataset.to_numpy()

X = data[:,0:17]
y = data[:,17:18]

# showing what is in the X array 
# X
```

```python
# showing what is in the y array or the labels 
y
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
selected = SelectKBest(chi2, k=3)
x_new = selected.fit_transform(X, y)
```

```python
filter = selected.get_support()
print("All features: ")
print(features)
print("Selected best 4: ")
count = 0
for filtered in filter:
    if filtered == True:
        print(features[count])
    count += 1
```

```python
from sklearn.model_selection import train_test_split

# creating the testing and training datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)

# creating the decision tree and fitting the train data to the decision tree

decision_tree = tree.DecisionTreeClassifier()
decision_tree = decision_tree.fit(X_train, y_train)
```

```python
import graphviz
dot_data = tree.export_graphviz(decision_tree, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("Predicting Delivery Method")
```

```python
# predicting the values given the current decision tree
predicted_values = decision_tree.predict(X_test)
predicted_values
```

```python
from sklearn.metrics import accuracy_score
accuracyScore = accuracy_score(y_test,predicted_values)

print("The accuracy score for this data is: " + str(accuracyScore), "%")
```

```python
from sklearn.model_selection import train_test_split

# creating the testing and training datasets
X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(x_new_tree, y, test_size=0.3, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.datasets import make_classification

clf = RandomForestClassifier(max_depth=2, random_state=0)
forest = clf.fit(X_train_tree, y_train_tree.ravel())
```

```python
predicted_tree = forest.predict(X_test_tree)
predicted_tree
```
